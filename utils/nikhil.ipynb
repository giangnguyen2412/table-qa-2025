```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error

#load the dataset
X_train = np.loadtxt('/content/X_train.csv')
Y_train = np.loadtxt('/content/Y_train.csv')
X_test = np.loadtxt('/content/X_test.csv')
Y_test = np.loadtxt('/content/Y_test.csv')

#Build neural network
model = Sequential()

# adding one  hidden layer with 5 neurons and sigmoid activation function
model.add(Dense(5, activation='sigmoid', input_shape=(X_train.shape[1],)))

# adding the output layer with one neuron
model.add(Dense(1))

# compiling the model with  Adam optimizer and  MSE loss
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')

# Train the model
history = model.fit(X_train, Y_train, epochs=1000, validation_split=0.2, verbose=1)

# evaluate on the Test
Y_pred = model.predict(X_test)
mse_test = mean_squared_error(Y_test, Y_pred)

print(f"Test MSE: {mse_test}")

#plotting the loss curve


plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()






```

```python
#4.

import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

# intialise the lists for every epochs
mse_history = []
r2_history = []

epochs = 1000

# Train the model
for epoch in range(epochs):

    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

# calculating the training loss
    mse_train = mean_squared_error(Y_train, y_pred_train)
    mse_test = mean_squared_error(Y_test, y_pred_test)

 # calculating R-squared
    r2_train = r2_score(Y_train, y_pred_train)
    r2_test = r2_score(Y_test, y_pred_test)

  # store the loss and accuracy for every epoch
    mse_history.append(mse_test)
    r2_history.append(r2_test)

    if epoch % 100 == 0:
        print(f'Epoch {epoch} - MSE: {mse_test}, R²: {r2_test}')

# plotting MSE for over epochs
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(epochs), mse_history, label='Test MSE Loss')
plt.title('Loss (MSE) vs. Number of Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()

# plotting R-squared over epochs
plt.subplot(1, 2, 2)
plt.plot(range(epochs), r2_history, label='Test R² Score', color='orange')
plt.title('Accuracy (R²) vs. Number of Epochs')
plt.xlabel('Epochs')
plt.ylabel('R² Score')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#5.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# Define different learning rates
learning_rates = [0.001, 0.01, 0.1, 0.5,1]

# list to store the results
r2_scores = []

for lr in learning_rates:

# creating neural network model for regression
    model = Sequential()
    model.add(Dense(5, input_dim=X_train.shape[1], activation='sigmoid'))
    model.add(Dense(1, activation='linear'))

# compile the model with varying learning rates
    optimizer = SGD(learning_rate=lr)
    model.compile(optimizer=optimizer, loss='mean_squared_error')

 # Train the model
    model.fit(X_train, Y_train, epochs=500, verbose=0)

 # evaluate the model on test data
    y_pred = model.predict(X_test)
    r2 = r2_score(Y_test, y_pred)

# store the R-squared
    r2_scores.append(r2)

    print(f"Learning Rate: {lr}, R² Score: {r2}")

# plot the results
plt.plot(learning_rates, r2_scores, marker='o')
plt.title('Accuracy (R²) vs. Learning Rate')
plt.xlabel('Learning Rate')
plt.ylabel('R² Score')
plt.xscale('log')
plt.grid(True)
plt.show()
```

Conclusion:
When the learning rate is low, the convergence in the training process takes places slowly but with less oscillations.

When the learning rate is optimal, the time needed for learning decreases and the oscillation decreases as well. The effect of high learning rates is the opposite.

Setting also large learning rate values can result in an instable training, wrong predictions and even make the training crash.

If you have observed that the learning rate will cause the model to jump around in response, choose a learning rate that is not too big or too small or a value of alpha approximately equal to 0.01 or 0.001 which is accepted in most architectures of neural networks.

```python
#7.b

from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Define activation functions
def build_model(activation_function):
    model = Sequential()
    model.add(Dense(5, input_dim=X_train.shape[1], activation=activation_function))
    model.add(Dense(1, activation='linear'))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

#activation Functions
activations = ['tanh', 'relu']
results = {}

for activation in activations:
   # create and train model
    model = build_model(activation)
    model.fit(X_train, Y_train, epochs=100, verbose=0)

   # evaluate the model on test data
    y_pred = model.predict(X_test)
    final_loss = mean_squared_error(Y_test, y_pred)

   # store results
    results[activation] = final_loss
#plot true Vs predicted values
    plt.figure(figsize=(6, 4))
    plt.plot(Y_test, y_pred, color='blue', label='Predicted')
    plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)], color='red', label='True line')
    plt.title(f"True vs Predicted using {activation} Activation Function")
    plt.xlabel('True Labels')
    plt.ylabel('Predicted Labels')
    plt.legend()
    plt.show()

# print final losses
for activation, loss in results.items():
    print(f"Final MSE Loss using {activation} activation: {loss:.4f}")









```

#8.a
 generally ,If the patience parameter is set too low i.e  5, early stopping will halt the training quickly if there's no improvement in the validation loss. While this can help prevent overfitting, it might also stop the training too soon, potentially before the model has had enough time to improve.

 With a higher patience value (e.g 20), the model will keep training for a longer period, allowing more learning over additional epochs. However, this increases the risk of overfitting, as the model may start memorizing the training data if the validation loss stops improving.

 #8.b

 As the patience value increases, both training and validation losses steadily decrease at all patience levels (5, 10, 20). This suggests that the model is taking advantage of the additional training time, allowing it to better fine-tune its parameters before early stopping starts .

from keras.regularizers import l2
from keras.callbacks import EarlyStopping

#define model with l2 regularisation
def build_model_with_l2(l2_value):
    model = Sequential()
    model.add(Dense(5, input_dim=X_train.shape[1], activation='tanh', kernel_regularizer=l2(l2_value)))
    model.add(Dense(1, activation='linear'))  # Output layer for regression
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# define l2 regularization strength and train the model
l2_value = 0.01
model_with_l2 = build_model_with_l2(l2_value)

# train the model with early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history_with_l2 = model_with_l2.fit(X_train, Y_train, epochs=1000, validation_split=0.2, callbacks=[early_stopping], verbose=0)

#plotting training Vs valdiation with l2 reguarlization
plt.plot(history_with_l2.history['loss'], label='Training Loss (L2 Regularization)')
plt.plot(history_with_l2.history['val_loss'], label='Validation Loss (L2 Regularization)')
plt.title('Training and Validation Loss with L2 Regularization')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()
# evaluating the model with L2 reguarlization
mse_loss_l2 = model_with_l2.evaluate(X_test, Y_test)
print(f"Final MSE Loss with L2 Regularization: {mse_loss_l2}")
